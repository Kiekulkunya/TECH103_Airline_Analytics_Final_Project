# Merge Data from main Data Set Airline Review and Location Latitude, Longitude, Distance_KM

```python
import pandas as pd
import numpy as np
import os

def intelligently_combine_dataframes(df1, df2):
    """
    Intelligently combine two dataframes to preserve all available information

    Args:
        df1 (pd.DataFrame): First input dataframe
        df2 (pd.DataFrame): Second input dataframe

    Returns:
        pd.DataFrame: Combined dataframe
    """
    # Print original columns and shapes
    print("File 1 Shape:", df1.shape)
    print("File 2 Shape:", df2.shape)
    print("\nColumns in First DataFrame:")
    print(df1.columns.tolist())
    print("\nColumns in Second DataFrame:")
    print(df2.columns.tolist())

    # Define comprehensive column order
    comprehensive_columns = [
        # Columns from the more detailed dataframe
        'No.', 'Airline Name', 'Overall_Rating', 'Review_Title', 'Review Date',
        'Verified', 'Review', 'Aircraft', 'Type Of Traveller', 'Seat Type',
        'Route', 'Date Flown', 'Seat Comfort', 'Cabin Staff Service',
        'Food & Beverages', 'Ground Service', 'Inflight Entertainment',
        'Wifi & Connectivity', 'Value For Money', 'Recommended',

        # Additional columns from the other dataframe
        'ReViennaw_Title_new', 'ReViennaw_new', 'Route_new',
        'Departure', 'Destination'
    ]

    # Prepare dataframes by adding missing columns
    def prepare_dataframe(df, column_order):
        """
        Prepare dataframe by adding missing columns

        Args:
            df (pd.DataFrame): Input dataframe
            column_order (list): Desired column order

        Returns:
            pd.DataFrame: Prepared dataframe
        """
        prepared_df = df.copy()

        # Add missing columns
        for col in column_order:
            if col not in prepared_df.columns:
                prepared_df[col] = np.nan

        # Reorder columns
        return prepared_df.reindex(columns=column_order)

    # Prepare both dataframes
    df1_prepared = prepare_dataframe(df1, comprehensive_columns)
    df2_prepared = prepare_dataframe(df2, comprehensive_columns)

    # Merge function to combine data from both sources
    def merge_row_data(row1, row2):
        """
        Merge data from two rows, prioritizing non-null values

        Args:
            row1 (pd.Series): First row of data
            row2 (pd.Series): Second row of data

        Returns:
            pd.Series: Combined row
        """
        merged_row = row1.copy()

        # Prioritize filling NaN values
        for col in row1.index:
            # If current value is NaN and other row has a value, use that
            if pd.isna(merged_row[col]) and pd.notna(row2[col]):
                merged_row[col] = row2[col]

        return merged_row

    # Combine rows based on 'No.' column
    # Create a dictionary to help matching and merging
    df1_dict = {row['No.']: row for _, row in df1_prepared.iterrows()}
    df2_dict = {row['No.']: row for _, row in df2_prepared.iterrows()}

    # Combine unique rows
    combined_rows = []
    all_nos = set(list(df1_dict.keys()) + list(df2_dict.keys()))

    for no in all_nos:
        row1 = df1_dict.get(no, pd.Series([np.nan] * len(comprehensive_columns), index=comprehensive_columns))
        row2 = df2_dict.get(no, pd.Series([np.nan] * len(comprehensive_columns), index=comprehensive_columns))

        # Merge the rows
        merged_row = merge_row_data(row1, row2)
        combined_rows.append(merged_row)

    # Create final combined dataframe
    combined_df = pd.DataFrame(combined_rows, columns=comprehensive_columns)

    # Verification
    print("\nMerge Verification:")
    print("Total Columns:", len(combined_df.columns))
    print("Columns in Combined DataFrame:", combined_df.columns.tolist())
    print("Total Rows:", len(combined_df))

    return combined_df

def combine_airline_reviews(file1, file2, output_file):
    """
    Combine two airline review CSV files

    Args:
        file1 (str): Path to first CSV file
        file2 (str): Path to second CSV file
        output_file (str): Path to output combined CSV file
    """
    # Read both CSV files with low_memory to handle mixed types
    try:
        df1 = pd.read_csv(file1, low_memory=False)
        df2 = pd.read_csv(file2, low_memory=False)
    except Exception as e:
        print(f"Error reading input files: {e}")
        return None

    # Combine dataframes
    combined_df = intelligently_combine_dataframes(df1, df2)

    # Save combined data
    try:
        # Ensure output is saved in the same directory
        if not output_file.startswith('/content/'):
            output_file = os.path.join('/content', output_file)

        combined_df.to_csv(output_file, index=False)
        print(f"\nCombined data saved to {output_file}")

        # Detailed verification
        print("\nData Verification:")
        print(f"Rows from File 1: {len(df1)}")
        print(f"Rows from File 2: {len(df2)}")
        print(f"Rows in Combined File: {len(combined_df)}")

        return combined_df

    except Exception as e:
        print(f"Error saving output file: {e}")
        return None

# Main execution
if __name__ == "__main__":
    # Specify input and output file paths
    input_file1 = '/content/Airline_reviews2.csv'
    input_file2 = '/content/Airline_reviews.csv'
    output_file = '/content/Airline_merge.csv'

    # Combine the dataframes
    df_merge = combine_airline_reviews(input_file1, input_file2, output_file) # Call the correct function to merge dataframes
```

# Data Preparation for Stage 2 - Merge file

```python
#!pip install geonamescache # Install the 'geonamescache' module before importing it.
import pandas as pd
import math
import geonamescache
import pandas as pd

# Update the URL to the raw CSV file
url = 'https://raw.githubusercontent.com/Kiekulkunya/TECH103_Airline_Analytics_Final_Project/refs/heads/main/Trip_Routes.csv'

# Define input_file
input_file = '/content/Trip_Routes.csv'  # Assign the correct file path to input_file

# Download the file from the URL to the specified input_file path
# Specify the encoding as 'latin-1' or 'ISO-8859-1' when reading the CSV from the URL
pd.read_csv(url, encoding='latin-1').to_csv(input_file, index=False)

try:
    df = pd.read_csv(input_file, encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(input_file, encoding='latin-1')

# 2. Build a city → (lat, lon) lookup via geonamescache
gc = geonamescache.GeonamesCache()
cities = gc.get_cities().values()
city_coords = {
    city['name'].title(): (float(city['latitude']), float(city['longitude']))
    for city in cities
}

# 3. Haversine distance helper
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in km
    φ1, φ2 = math.radians(lat1), math.radians(lat2)
    Δφ = math.radians(lat2 - lat1)
    Δλ = math.radians(lon2 - lon1)
    a = math.sin(Δφ/2)**2 + math.cos(φ1)*math.cos(φ2)*math.sin(Δλ/2)**2
    return 2 * R * math.atan2(math.sqrt(a), math.sqrt(1 - a))

# 4. Lookup functions for lat/lon
def lookup_lat(name):
    return city_coords.get(name, (pd.NA, pd.NA))[0]

def lookup_lon(name):
    return city_coords.get(name, (pd.NA, pd.NA))[1]

# 5. Create the new latitude/longitude columns
df['depart_latitude']  = df['Departure'].apply(lookup_lat)
df['depart_longitude'] = df['Departure'].apply(lookup_lon)
df['des_latitude']     = df['Destination'].apply(lookup_lat)
df['des_longitude']    = df['Destination'].apply(lookup_lon)

# 6. Compute distance (km) if coordinates are available
df['Distance_km'] = df.apply(
    lambda r: haversine(r.depart_latitude, r.depart_longitude,
                        r.des_latitude, r.des_longitude)
              if pd.notna(r.depart_latitude) and pd.notna(r.des_latitude)
              else pd.NA,
    axis=1
)

# 7. Save your results
output_file = '/content/Trip_Routes2.csv'
df.to_csv(output_file, index=False)

print(f"✅ Trip Route 2 file saved to {output_file}")
```

# Create Flight Classification by assigning Short, Medium, Long Haul Route

```python
import pandas as pd
import re

# auto-install airportsdata if you don't have it:
# !pip install airportsdata

from airportsdata import load

# 1. Load your CSV
input_file = '/content/airline_new.csv'
df = pd.read_csv(input_file)

# 2. Name of the column with your raw routes (e.g. "FRA/ZUR" or "Zurich-Vienna-Zurich")
raw_col = 'ReViennaw_new'   # ← change if yours is called something else

# 3. Load a full IATA→airport info mapping
#    (includes 'city' keys for every standard 3-letter code)
airports = load('IATA')

# 4. Parser: split on hyphens/slashes/dashes, map codes→cities, collapse round-trips
def parse_route(route):
    if not isinstance(route, str) or not route.strip():
        return pd.Series([pd.NA, pd.NA, pd.NA],
                         index=['Route_new','Departure','Destination'])

    # split on -, /, or –
    stops = re.split(r'[-/–]', route)
    stops = [s.strip() for s in stops if s.strip()]

    # normalize each stop: if it's a 3-letter IATA code, fetch its city
    cities = []
    for stop in stops:
        code = stop.upper()
        if code in airports:
            cities.append(airports[code]['city'])
        else:
            # fallback: title-case whatever string it was
            cities.append(stop.title())

    if len(cities) < 2:
        return pd.Series([pd.NA, pd.NA, pd.NA],
                         index=['Route_new','Departure','Destination'])

    dep = cities[0]
    # pick the last city ≠ the departure (to collapse round trips)
    dest = next((c for c in reversed(cities) if c != dep), cities[-1])
    route_new = f"{dep} to {dest}"
    return pd.Series([route_new, dep, dest],
                     index=['Route_new','Departure','Destination'])

# 5. Ensure the new columns exist (so we can mask them)
for col in ['Route_new','Departure','Destination']:
    if col not in df.columns:
        df[col] = pd.NA

# 6. Build mask: rows where we still need to fill them
mask = (
    df['Route_new'].isna()
    | df['Route_new'].astype(str).str.strip().eq('')
)

# 7. Apply parsing only on the missing ones
df.loc[mask, ['Route_new','Departure','Destination']] = (
    df.loc[mask, raw_col]
      .apply(parse_route)
)

# 8. (Optional) now you can compute distance+haul type as before…

# 9. Save out
output_file = '/content/airline_clean.csv'
df.to_csv(output_file, index=False)
print(f"✔️  Cleaned routes written to {output_file}")


distant_count = df['Distance_km'].nunique() # Changed 'distance_km' to 'Distance_km' to match column name
print(f"Total number of distance: {distant_count}") # Changed 'distance_count' to 'distant_count' to match variable name
```

```python
# auto-install airportsdata and geonamescache if you don't have them:
# !pip install airportsdata geonamescache

import pandas as pd
import re
import numpy as np
from airportsdata import load as load_airports
import geonamescache
from math import radians, sin, cos, atan2, sqrt

# 1. Load your CSV
input_file = '/content/airline_clean.csv'
df = pd.read_csv(input_file)

# 2. Name of the column with your raw routes
raw_col = 'ReViennaw_new'

# 3. Load IATA→airport info (for coordinate fallback) and city cache
airports = load_airports('IATA')
gc = geonamescache.GeonamesCache()
city_coords = {
    info['name']:(float(info['latitude']), float(info['longitude']))
    for info in gc.get_cities().values()
}

def get_coords(place):
    """Return (lat,lon) for a city name or IATA code."""
    if not isinstance(place, str):
        return (np.nan, np.nan)
    up = place.upper()
    # IATA code?
    if up in airports:
        return (airports[up]['lat'], airports[up]['lon'])
    # city lookup
    return city_coords.get(place.title(), (np.nan, np.nan))

# 4. Parser: split on hyphens/slashes/dashes, map codes→cities, extract Route_new
def parse_route(route):
    if not isinstance(route, str) or not route.strip():
        return pd.Series([pd.NA, pd.NA, pd.NA],
                         index=['Route_new','Departure','Destination'])
    stops = re.split(r'[-/–]', route)
    stops = [s.strip() for s in stops if s.strip()]
    cities = []
    for stop in stops:
        code = stop.upper()
        if code in airports:
            cities.append(airports[code]['city'])
        else:
            cities.append(stop.title())
    if len(cities) < 2:
        return pd.Series([pd.NA, pd.NA, pd.NA],
                         index=['Route_new','Departure','Destination'])
    dep = cities[0]
    dest = next((c for c in reversed(cities) if c != dep), cities[-1])
    return pd.Series([f"{dep} to {dest}", dep, dest],
                     index=['Route_new','Departure','Destination'])

# 5. Ensure the new columns exist
for col in ['Route_new','Departure','Destination']:
    if col not in df.columns:
        df[col] = pd.NA

# 6. Fill only missing Route_new / Departure / Destination
mask = df['Route_new'].isna() | df['Route_new'].astype(str).str.strip().eq('')
df.loc[mask, ['Route_new','Departure','Destination']] = (
    df.loc[mask, raw_col].apply(parse_route)
)

# 7. Compute Distance_km where missing, using Haversine on Departure & Destination
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius km
    φ1, φ2 = map(radians, (lat1, lat2))
    Δφ = radians(lat2 - lat1)
    Δλ = radians(lon2 - lon1)
    a = sin(Δφ/2)**2 + cos(φ1)*cos(φ2)*sin(Δλ/2)**2
    return 2*R*atan2(sqrt(a), sqrt(1-a))

# initialize Distance_km column if absent
if 'Distance_km' not in df.columns:
    df['Distance_km'] = np.nan

mask_dist = df['Distance_km'].isna() & df['Route_new'].notna()
for idx in df[mask_dist].index:
    dep, dest = df.at[idx, 'Departure'], df.at[idx, 'Destination']
    lat1, lon1 = get_coords(dep)
    lat2, lon2 = get_coords(dest)
    if pd.notna(lat1) and pd.notna(lat2):
        df.at[idx, 'Distance_km'] = haversine(lat1, lon1, lat2, lon2)

# 8. Save your updated dataset
output_file = '/content/airline_clean2.csv'
df.to_csv(output_file, index=False)
print(f"✔️  Filled missing distances and saved to {output_file}")

distant_count = df['Distance_km'].nunique() # Changed 'distance_km' to 'Distance_km' to match column name
print(f"Total number of distance: {distant_count}") # Changed 'distance_count' to 'distant_count' to match variable nam
```

# Applying GPT Research outcomes to Flight Type for filling missing data

```python
import pandas as pd
import numpy as np
import re

def create_airline_classification_map():
    """
    Create a comprehensive classification map for airlines

    Returns:
        dict: Mapping of airlines to their haul classification
    """
    return {
        # Long Haul Airlines (Intercontinental routes)
        "American Airlines": "Long Haul",
        "Delta Air Lines": "Long Haul",
        "United Airlines": "Long Haul",
        "Air Canada": "Long Haul",
        "British Airways": "Long Haul",
        "Lufthansa": "Long Haul",
        "Air France": "Long Haul",
        "Emirates": "Long Haul",
        "Etihad Airways": "Long Haul",
        "Qatar Airways": "Long Haul",
        "Singapore Airlines": "Long Haul",
        "Qantas": "Long Haul",
        "Cathay Pacific": "Long Haul",
        "Turkish Airlines": "Long Haul",
        "Air India": "Long Haul",

        # Medium Haul Airlines (Continental routes)
        "JetBlue Airways": "Medium Haul",
        "Ryanair": "Medium Haul",
        "easyJet": "Medium Haul",
        "AirAsia": "Medium Haul",
        "All Nippon Airways (ANA)": "Medium Haul",
        "KLM Royal Dutch Airlines": "Medium Haul",

        # Short Haul Airlines (Regional routes)
        "Southwest Airlines": "Short Haul",
        "SpiceJet": "Short Haul",
        "IndiGo": "Short Haul",
        "Aer Lingus": "Short Haul",
        "EasyJet": "Short Haul",
        "Ryanair": "Short Haul"
    }

def classify_flight_types(input_file, output_file):
    """
    Classify flight types using multiple methods

    Args:
        input_file (str): Path to input CSV file
        output_file (str): Path to output CSV file

    Returns:
        pd.DataFrame: Processed DataFrame
    """
    # 1. Load Data
    try:
        df = pd.read_csv(input_file, low_memory=False)
    except Exception as e:
        print(f"Error reading input file: {e}")
        return None

    # Ensure Flight_type column exists
    if 'Flight_type' not in df.columns:
        df['Flight_type'] = 'Unknown'

    # 2. Distance-based Categorization
    try:
        # Define distance thresholds
        low, high = 1670, 6022

        # Mask for rows with valid distance
        mask_dist = df['Distance_km'].notna()

        # Categorize based on distance
        df.loc[mask_dist, 'Flight_type'] = pd.cut(
            df.loc[mask_dist, 'Distance_km'],
            bins=[-1, low, high, np.inf],
            labels=['Short_Haul', 'Medium_Haul', 'Long_Haul']
        )
    except Exception as e:
        print(f"Error in distance-based categorization: {e}")

    # 3. Classification-based Filling
    try:
        # Create classification map
        classification_map = create_airline_classification_map()

        # Mask for rows still marked as Unknown
        mask_unknown = df['Flight_type'] == 'Unknown'

        # Function to map airline to flight type
        def map_airline_to_flight_type(airline):
            # First check exact match
            if airline in classification_map:
                return classification_map[airline]

            # Then try partial matching
            for key, value in classification_map.items():
                if key.lower() in airline.lower():
                    return value

            # Default to Short Haul if no match
            return 'Short_Haul'

        # Apply mapping to unknown rows
        df.loc[mask_unknown, 'Flight_type'] = df.loc[mask_unknown, 'Airline Name'].apply(map_airline_to_flight_type)
    except Exception as e:
        print(f"Error in classification-based filling: {e}")

    # 4. Print Distribution
    print("\nFlight Type Distribution:")
    distribution = df['Flight_type'].value_counts(dropna=False)
    print(distribution)

    # 5. Additional Detailed Counts
    print("\nDetailed Flight Type Counts:")
    print("Unclassified:", df[df['Flight_type'] == 'Unknown'].shape[0])
    print("Short Haul:", df[df['Flight_type'] == 'Short_Haul'].shape[0])
    print("Medium Haul:", df[df['Flight_type'] == 'Medium_Haul'].shape[0])
    print("Long Haul:", df[df['Flight_type'] == 'Long_Haul'].shape[0])

    # 6. Save Processed DataFrame
    try:
        df.to_csv(output_file, index=False)
        print(f"\n✅ All missing Flight_type values filled → {output_file}")
    except Exception as e:
        print(f"Error saving output file: {e}")

    return df

# Main execution
if __name__ == "__main__":
    input_file = '/content/airline_clean2.csv'
    output_file = '/content/airline_classified.csv'

     # Process and classify flight types
    df_classification = classify_flight_types(input_file, output_file)
```

```python
import os
import re
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer

def load_dataframe(file_paths=None):
    """
    Load DataFrame with multiple fallback options

    :param file_paths: List of potential file paths to load
    :return: Loaded DataFrame
    """
    # Default file paths to try
    default_paths = [
        '/content/Airline_embedded.pkl',
        '/content/drive/MyDrive/TECH103_Airline_Semantic_Search/Airline_embedded.pkl',
        '/content/drive/MyDrive/TECH103_Airline_Semantic_Search/entire_data.pkl',
        'Airline_embedded.pkl',
        'entire_data.pkl'
    ]

    # Merge user-provided paths with default paths
    if file_paths:
        paths_to_try = file_paths + default_paths
    else:
        paths_to_try = default_paths

    # Try loading from various paths
    for path in paths_to_try:
        try:
            print(f"🔍 Attempting to load DataFrame from: {path}")
            df = pd.read_pickle(path)
            print(f"✅ Successfully loaded {len(df)} rows from {path}")

            # Ensure required columns exist
            required_columns = ['Airline Name', 'Overall_Rating', 'Recommended', 'combined']
            missing_columns = [col for col in required_columns if col not in df.columns]

            if missing_columns:
                print(f"⚠️ Warning: Missing columns: {missing_columns}")
                continue

            return df

        except FileNotFoundError:
            print(f"❌ File not found: {path}")
        except Exception as e:
            print(f"❌ Error loading {path}: {e}")

    # If no file is found, raise an error
    raise FileNotFoundError("Could not load DataFrame from any of the specified paths")

def prepare_search_query(query):
    """
    Prepare and extract search components from the query

    :param query: Input search query
    :return: Dictionary of search components
    """
    # Convert to lowercase and strip
    query = query.lower().strip()

    # Initialize search components
    search_components = {
        'rating': None,
        'flight_type': None,
        'departure': None,
        'destination': None,
        'recommended': None,
        'keywords': []
    }

    # Extract rating
    rating_match = re.search(r'rating\s*(?:greater\s*than|>)\s*(\d+(?:\.\d+)?)', query)
    if rating_match:
        search_components['rating'] = float(rating_match.group(1))

    # Extract flight type
    flight_type_matches = {
        'long haul': ['long', 'long haul', 'longhaul'],
        'short haul': ['short', 'short haul', 'shorthaul']
    }
    for type_key, type_variations in flight_type_matches.items():
        if any(var in query for var in type_variations):
            search_components['flight_type'] = type_key

    # Extract departure
    departure_match = re.search(r'from\s*([^t]+)(?:\s*to|$)', query)
    if departure_match:
        search_components['departure'] = departure_match.group(1).strip()

    # Extract destination
    destination_match = re.search(r'to\s*(.+)$', query)
    if destination_match:
        search_components['destination'] = destination_match.group(1).strip()

    # Check for recommended
    if 'recommended' in query:
        search_components['recommended'] = 'yes'

    # Extract remaining keywords
    search_components['keywords'] = [
        word for word in query.split()
        if word not in ['airlines', 'with', 'from', 'to', 'is', 'the']
    ]

    return search_components

def advanced_search(df, search_components):
    """
    Perform advanced search based on extracted components

    :param df: Input DataFrame
    :param search_components: Dictionary of search components
    :return: Filtered DataFrame
    """
    # Start with the full DataFrame
    filtered_df = df.copy()

    # Filter by rating
    if search_components['rating'] is not None:
        filtered_df = filtered_df[filtered_df['Overall_Rating'] > search_components['rating']]

    # Filter by flight type
    if search_components['flight_type']:
        filtered_df = filtered_df[
            filtered_df['Flight_type'].str.contains(search_components['flight_type'], case=False, na=False)
        ]

    # Filter by departure
    if search_components['departure']:
        filtered_df = filtered_df[
            filtered_df['Departure'].str.contains(search_components['departure'], case=False, na=False)
        ]

    # Filter by destination
    if search_components['destination']:
        filtered_df = filtered_df[
            filtered_df['Destination'].str.contains(search_components['destination'], case=False, na=False)
        ]

    # Filter by recommended
    if search_components['recommended']:
        filtered_df = filtered_df[filtered_df['Recommended'] == search_components['recommended']]

    # Fallback text search if no other filters applied
    if len(filtered_df) == 0 and search_components['keywords']:
        # Search across multiple columns
        search_columns = [
            'Airline Name', 'Review', 'Route',
            'Departure', 'Destination', 'Flight_type',
            'combined'
        ]

        # Create search condition
        search_condition = pd.Series([False] * len(df))
        for keyword in search_components['keywords']:
            column_conditions = [
                df[col].str.contains(keyword, case=False, na=False)
                for col in search_columns
            ]
            search_condition |= pd.concat(column_conditions, axis=1).any(axis=1)

        filtered_df = df[search_condition]

    # Sort by overall rating
    filtered_df = filtered_df.sort_values('Overall_Rating', ascending=False)

    return filtered_df

def main():
    # Load DataFrame
    df = load_dataframe()

    # Diagnostic information function
    def diagnostic_dataframe_info():
        """
        Provide comprehensive diagnostic information about the DataFrame
        """
        print("\n=== DataFrame Diagnostic Information ===")

        # Basic DataFrame info
        print(f"Total rows: {len(df)}")

        # Recommended column analysis
        print("\n=== Recommended Column Analysis ===")
        print("Unique values in 'Recommended' column:")
        print(df['Recommended'].value_counts())

        # Recommended column boolean check
        print("\nBoolean check of 'Recommended' column:")
        print(df['Recommended'].unique())

        # Column names
        print("\n=== Columns ===")
        print(df.columns.tolist())

    # Print diagnostic information
    diagnostic_dataframe_info()

    # Interactive search function
    def interactive_cli_search():
        """
        Robust command-line interactive search interface
        """
        print("🛫 Airline Semantic Search")
        print("Type 'exit' to quit")

        while True:
            try:
                query = input("\n 🔎 🛫 Enter your search query: ").strip()

                if query.lower() in ['exit', 'quit']:
                    print(" ✨ Thank you for using Airline Search!")
                    break

                # Skip empty queries
                if not query:
                    continue

                # Prepare search query
                search_components = prepare_search_query(query)

                # Perform advanced search
                results = advanced_search(df, search_components)

                if len(results) > 0:
                    print(f"\n=== Search Results ({len(results)} found) ===")
                    for i, (_, row) in enumerate(results.head(10).iterrows(), 1):
                        print(f"\n{i}. Airline: {row['Airline Name']}")
                        print(f"   Overall Rating: {row['Overall_Rating']:.2f}")
                        print(f"   Recommended: {row['Recommended']}")
                        print(f"   Flight Type: {row['Flight_type']}")
                        print(f"   Route: {row['Departure']} to {row['Destination']}")
                        print(f"   Review (excerpt): {row['Review'][:200]}...")
                else:
                    print("No results found.")

            except Exception as e:
                print(f"An unexpected error occurred: {e}")
                import traceback
                traceback.print_exc()
                continue

    # Run interactive search
    interactive_cli_search()

if __name__ == "__main__":
    main()
```

```python
from tiktoken import get_encoding

def chunk_text(text: str, chunk_size: int = 500, overlap: int = 50) -> list[str]:
    """
    Split `text` into token-level chunks of up to `chunk_size` tokens,
    with `overlap` tokens of context between successive chunks.
    """
    enc = get_encoding("cl100k_base")
    tokens = enc.encode(text)
    chunks = []
    for i in range(0, len(tokens), chunk_size - overlap):
        chunk = tokens[i: i + chunk_size]
        chunks.append(enc.decode(chunk))  # Corrected indentation here to align with for loop
    return chunks  # Corrected indentation here to align with function definition
```

```python
from tiktoken import get_encoding
from qdrant_client import QdrantClient
from qdrant_client.http import models as rest
from sentence_transformers import SentenceTransformer
import pandas as pd

# 1) Chunking helper
def chunk_text(text: str, chunk_size: int = 500, overlap: int = 50) -> list[str]:
    enc = get_encoding("cl100k_base")
    tokens = enc.encode(text)
    chunks = []
    for i in range(0, len(tokens), chunk_size - overlap):
        chunk = tokens[i : i + chunk_size]
        chunks.append(enc.decode(chunk))
    return chunks

# 2) Load your DataFrame
df = pd.read_pickle('/content/drive/MyDrive/TECH103_Airline_Semantic_Search/entire_data.pkl')

# 3) Filter & sample 100 rows
df_filtered = df.dropna(subset=['Review'])              # require a non-null review
df_sample   = df_filtered.sample(n=100, random_state=42)  # reproducible 100-row random sample

# 4) Initialize Qdrant + Embedder
try:
    qdrant  = QdrantClient(path='/content/drive/MyDrive/TECH103_Airline_Semantic_Search/qdrant_data')
    embedder = SentenceTransformer('all-MiniLM-L6-v2')
except RuntimeError as e:
    if "already accessed by another instance" in str(e):
        print("⚠️ Qdrant already initialized—reusing existing client.")
    else:
        raise

# 5) (Re)create the collection
qdrant.recreate_collection(
    collection_name="airline_reviews",
    vectors_config=rest.VectorParams(size=384, distance=rest.Distance.COSINE)
)

# 6) Chunk → Embed → Collect points (on the 100-row sample)
points = []
uid    = 0
for _, row in df_sample.iterrows():
    for chunk in chunk_text(row["Review"]):
        vec = embedder.encode(chunk)
        points.append(
            rest.PointStruct(
                id=uid,
                vector=vec.tolist(),
                payload={
                    "chunk": chunk,
                    "airline":       row["Airline Name"],
                    "overall_rating":row["Overall_Rating"],
                    "recommended":   row["Recommended"],
                    "review_date":   row["Review Date"],
                    "route":         row["Route_new"],
                    "departure":     row["Departure"],
                    "destination":   row["Destination"],
                    "flight_type":   row["Flight_type"],
                    "seat_type":     row["Seat Type"],
                    "aircraft":      row["Aircraft"],
                    "review_title":  row["Review_Title"],
                    "combined":  row["combined"]
                }
            )
        )
        uid += 1

# 7) Batch upsert into Qdrant
qdrant.upsert(
    collection_name="airline_reviews",
    points=points
)

print(f"✅ Upserted {len(points)} chunks (from 100 reviews) into Qdrant!")
```

```python
# This cell imports gradio and defines the generate_answer function
import gradio as gr
import openai


def generate_answer(query):
    try:

        sources = query_qdrant(query, qdrant)
        ##sources = query_qdrant(query, client)
    except Exception as e:
        # If something goes wrong (e.g., Qdrant is not running), return a fallback message
        sources = [{"error": f"Error retrieving sources: {str(e)}"}]
        print(f"Error retrieving sources: {str(e)}") # Print the error for debugging

    # Step 2: Prepare the prompt for the language model
    # Includes the user's question and the context retrieved from the vector database
    # Ensure the search function or logic returning context is defined and accessible
    context_text = ""
    if sources and 'error' not in sources[0]:
         # Assuming sources is a list of dictionaries/objects with a 'content' key
         context_text = "\n".join([str(s.get('content', '')) for s in sources if s and s.get('content')])
    elif sources and 'error' in sources[0]:
         context_text = f"Error retrieving context: {sources[0]['error']}"
         print(context_text) # Print context error

    prompt = f"""
    Based on the following query from a user and provided context (if available), please generate a concise answer.
    Focus on the original query and the response given in the context.
    Remove special characters and newlines (/n).
    Format the output cleanly and long, potentially using bullet points.
    Use emojis for support text where appropriate.
    Start directly with the answer — no salutations.

    ###########
    query:
    "{query}"

    ########

    context:
    "{sources}"
    #####

    Return in Markdown format.
    """

    # Step 3: Send the prompt to the OpenRouter-compatible LLM (Qwen model)
    # Ensure open_router_client is initialized (e.g., in a preceding cell)
    if 'open_router_client' not in globals() or open_router_client is None:
        print("Error: 'open_router_client' is not initialized.")
        yield "Sorry, the AI model is not available due to a configuration error."
        return

    try:
        stream = open_router_client.chat.completions.create(
            model="qwen/qwen3-8b:free",  # A free-to-use large language model hosted on OpenRouter
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                },
            ],
            stream=True,  # Enable streaming response
        )

        # Step 4: Stream and yield the generated content chunk by chunk
        full_response = ""
        for chunk in stream:
            if chunk.choices[0].delta.content is not None:
                content = chunk.choices[0].delta.content
                full_response += content

                # Yield lets us return partial results as they're received (for real-time feedback)
                yield full_response

    except Exception as e:
        print(f"Error generating answer from LLM: {e}")
        yield f"Sorry, there was an error generating the answer: {str(e)}"

```

# Airport Visualization from Airline Reviews observation. This is a a Real World Airport Trafficking

```python
import pandas as pd
import folium
import numpy as np
from IPython.display import display, HTML

# Load merged dataset
df = pd.read_csv('/content/flight_weather_radar.csv')

# Compute departure & destination counts
dep = (
    df.groupby(['Departure', 'depart_latitude', 'depart_longitude'])
      .size().reset_index(name='dep_count')
      .rename(columns={
          'Departure': 'Airport',
          'depart_latitude': 'Latitude',
          'depart_longitude': 'Longitude'
      })
)
dest = (
    df.groupby(['Destination', 'des_latitude', 'des_longitude'])
      .size().reset_index(name='dest_count')
      .rename(columns={
          'Destination': 'Airport',
          'des_latitude': 'Latitude',
          'des_longitude': 'Longitude'
      })
)

# Merge counts & compute total
counts = pd.merge(dep, dest, on=['Airport', 'Latitude', 'Longitude'], how='outer').fillna(0)
counts['total_count'] = counts['dep_count'] + counts['dest_count']

# Determine percentile thresholds
p35, p55, p75, p90 = np.percentile(counts['total_count'], [35, 55, 75, 90])

def assign_color(cnt):
    if cnt >= p90:
        return 'red'
    elif cnt >= p75:
        return 'pink'
    elif cnt >= p55:
        return 'lightgreen'
    elif cnt >= p35:
        return 'lightblue'
    else:
        return 'gray'

counts['color'] = counts['total_count'].apply(assign_color)

# Build Folium map with airplane icons
center = [counts['Latitude'].mean(), counts['Longitude'].mean()]
m = folium.Map(location=center, zoom_start=2)

# Add markers with FontAwesome airplane icons
for _, row in counts.iterrows():
    folium.Marker(
        location=(row['Latitude'], row['Longitude']),
        icon=folium.Icon(icon='plane', prefix='fa', color=row['color']),
        popup=f"{row['Airport']}: {int(row['total_count'])} flights"
    ).add_to(m)

# Add a legend
legend_html = '''
<div style="
    position: fixed;
    bottom: 50px; left: 70px; width: 180px; height: 140px;
    border:2px solid grey; background:white; z-index:9999;
    font-size:14px; padding: 10px;
">
<b>Airline Traffic from review </b><br>
<i class="fa fa-plane" style="color:red; margin-right:5px;"></i>Super busy Airports<br>
<i class="fa fa-plane" style="color:pink; margin-right:5px;"></i>Busy Airports<br>
<i class="fa fa-plane" style="color:lightgreen; margin-right:5px;"></i>Normal traffic<br>
<i class="fa fa-plane" style="color:lightblue; margin-right:5px;"></i>Low traffic<br>
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# Display inline
display(HTML(m._repr_html_()))

# Save map to HTML file
map_path = '/content/airport_busyness_map.html'
m.save(map_path)
print(f"Plane icon map saved to: {map_path}")
```
